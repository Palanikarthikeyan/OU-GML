```python
BFS
| -> Explore all the neighbors of node - each level
| -> queue - keep all the visiting nodes

(City1) ----- (City2)
    |           |
(City3)  -----( City4)

Adjancy matrix
   C1   C2  C3 C4
C1  0    1   1  0
C2  1    0   0  1
C3
C4
initialize start node = 1  ->Queue - [ 1 ] ;visited - 1
next visiting node 1  
                

    
```

```python
G = nx.Graph()
G.add_edges_from([('A','B'),('A','C'),('B','D'),('B','E'),('C','F'),('C','G')])

def bfs_search(graph,node):
    visited,queue = [node],[node]
    while queue:
        node =queue.pop(0)
        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.append(neighbor)
                queue.append(neighbor)
    return visited
    
bfs_search(G,'A')
#bfs_search(G,'B')
```

```python
# DFS - Depth First Search
# backtracking - stack 

start root node -> Visited Node => visit an unvisited neighbors  - repeat until 
                                   there are no unvisited neighbors
G = nx.Graph()
'A' -> 'B','C' 
'B' -> 'D','E'
'C' -> 'F'
'D' -> 
'E' -> 'F'
'F' ->

A
|__ B,C
    | |___F 
    |__D,E
          |__F
```

```python
G = nx.Graph()
G.add_edges_from([('A','B'),('A','C'),('B','D'),('B','E'),('C','F'),('C','G')])

visited = []
def f1(visited,graph,node):
    if node not in visited:
        visited.append(node)
        for neighbor in graph[node]:
            visited=f1(visited,graph,neighbor)
    return visited

print(visited)
f1(visited,G,'A')
```
