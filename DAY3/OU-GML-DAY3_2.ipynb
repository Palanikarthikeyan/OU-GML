```python
import networkx as nx
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
Gobj = nx.erdos_renyi_graph(10,0.3)
nx.draw(Gobj,with_labels=True)
```

```python
nx.draw_networkx(Gobj,pos=nx.spring_layout(Gobj,seed=0),node_size=600,font_size=13,font_color='white')
```

```python
def random_walk(start_node,length):
    walk = [str(start_node)] # starting node
    for i in range(length):
        neighbors = [var for var in Gobj.neighbors(start_node)]
        next_node = np.random.choice(neighbors,1)[0]
        walk.append(str(next_node))
        start_node = next_node
    return walk
```

```python
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# load datasets
iris = load_iris()
X = iris.data
Y = iris.target

x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.4,random_state=42)
```

```python
# create model (Random forest model)
model = RandomForestClassifier(n_estimators=100,random_state=42)
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
accuracy_score(y_test,y_pred)
```

```python
features = iris.feature_names
importances = model.feature_importances_
indices = importances.argsort()[::-1]
```

```python
import networkx as nx
from karateclub import Graph2Vec
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

Gobj = nx.karate_club_graph()
```

```python
from karateclub import Node2Vec

# train node2vec
model = Node2Vec()
model.fit(Gobj)
embeddings = model.get_embedding()

# prepare labels
labels = [0 if Gobj.nodes[var]['club'] == 'Mr. Hi' else 1 for var in Gobj.nodes()]

x_train,x_test,y_train,y_test = train_test_split(embeddings,labels,test_size=0.4,random_state=42)

### Train Random Forest Classifier

obj = RandomForestClassifier(n_estimators=100,random_state=42)
obj.fit(x_train,y_train)

### Evaluate 
y_pred = obj.predict(x_test)
print(classification_report(y_test,y_pred))
```
