```python
# 1. load the data (graph) - load the graph
import networkx as nx
Gobj = nx.karate_club_graph()

# 2. prepare node parameters - labels 
for var in Gobj.nodes:
    Gobj.nodes[var]['y'] = 0 if Gobj.nodes[var]['club'] == 'Mr. Hi' else 1

# Mr. Hi -> 0    ; officer ->1

# 3. pytoch gemetric data
import torch.nn.functional as F
for var in Gobj.nodes:
    Gobj.nodes[var]['x'] = F.one_hot(torch.tensor(var),num_classes=Gobj.number_of_nodes()).float()
```

```python
Gobj.nodes
```

```python
# 4. build the model

from torch_geometric.utils import from_networkx
data = from_networkx(Gobj) # Geometric Data - this is input to model

from torch_geometric.nn import GCNConv
import torch.nn as nn

class GCN(nn.Module):
    def __init__(self):
        super().__init__() # nn.Module.__init__(self)
        self.conv1 = GCNConv(data.num_node_features,16)
        self.conv2 = GCNConv(16,2)
    def forward(self,x,edge_index):
        x = self.conv1(x,edge_index).relu()
        x = self.conv2(x,edge_index)
        return x
```

```python
# 5. train the model
model = GCN()
optimizer = torch.optim.Adam(model.parameters())
x,edge_index,y = data.x,data.edge_index,data.y
```
